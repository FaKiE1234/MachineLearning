```python
import tensorflow as tf
import numpy as np
```

```python
x = []
y = []

for _ in range(10000):
    num = np.random.randint(0, 101)
    x.append( num )
    y.append( num % 2 )
    
    
```

```python
## inferance function

model = tf.keras.Sequential()
model.add(  tf.keras.layers.Dense(input_shape=(1,), units=64, activation='sigmoid')   ) ##input layer + 64 nuran layer
model.add(  tf.keras.layers.Dense(units=32, activation='sigmoid')   )
model.add(  tf.keras.layers.Dense(units=16, activation='sigmoid')   )
model.add(  tf.keras.layers.Dense(units=8, activation='sigmoid')   )
model.add(  tf.keras.layers.Dense(units=1, activation='sigmoid')   ) ## output layer
```

```python
model.compile('adam', 'binary_crossentropy')
```

```python
model.fit(x, y, epochs=100)
```

```python
model.save('modelJP.h5')
```
